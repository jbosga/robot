```python
import re
```

```python
with open('data/coco_labels.txt', 'r') as f:
    labels = f.read()
```

```python
print(labels)
```

```python
names = re.findall('("\w+")|("\w+\s\w+")', labels)
names = [name[0]+name[1] for name in names]
names = [name.replace('"', '') for name in names]
```

```python
ids = re.findall('id: \d+', labels)
ids = [idx.replace('id: ', '') for idx in ids]
```

```python
new_labels = dict(zip(ids, names))
```

```python
import json
```

```python
with open('data/80_coco_labels.json', 'w') as f:
    f.write(json.dumps(new_labels))
```
